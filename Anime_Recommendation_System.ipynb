**BASIC CLEANING OF DATA**

```python
df = df[['Name', 'Genres', 'sypnopsis', 'type']]
df.head()
```

Generating a WordCloud to see the most frequently used keywords in Genres

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
words = ""
for i in df['Genres']:
    i = str(i)
    separate = i.split()
    for j in range(len(separate)):
        separate[j] = separate[j].lower()

    words+= " ".join(separate)+" "

stop_words = set(STOPWORDS)

wordcloud = WordCloud(width = 800, height = 800,
                background_color ='black',
                stopwords = stop_words,
                min_font_size = 10).generate(words)
```

```python
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```

**CONTENT BASED FILTERING**

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
indices = pd.Series(df.index, index = df['Name'])
print(indices)
```

Similarity based on the plot(sypnopsis) of the anime

```python
tfidf = TfidfVectorizer(stop_words = 'english')
sypnopsis_matrix = tfidf.fit_transform(df['sypnopsis'])
sypnopsis_matrix.shape
```

```python
plot_similarity = cosine_similarity(sypnopsis_matrix)
```

```python
def recommendations(title, matrix):
  ind = indices[title]
  scores = list(enumerate(matrix[ind]))
  scores = sorted(scores, key=lambda x:x[1], reverse = True)
  scores = scores[1:11]

  anime_indices = [k[0] for k in scores]

  return df['Name'].iloc[anime_indices]
```

```python
recommendations('One Punch Man', plot_similarity)
```

Similarity based on genres

```python
count = CountVectorizer(stop_words = 'english')
genre_matrix = count.fit_transform(df['Genres'])
```

```python
genre_similarity = cosine_similarity(genre_matrix)
```

```python
recommendations('One Punch Man', genre_similarity)
```

```python
tfidf2 = TfidfVectorizer(stop_words = 'english')
name_matrix = tfidf2.fit_transform(df['Name'])
name_similarity = cosine_similarity(name_matrix)
```

```python
recommendations('One Punch Man', name_similarity)
```

Modified function to evaluate similarity across multiple benchmarks

```python
def recommendations_modified(title, matrix1, matrix2):
  ind = indices[title]

  scores1 = list(enumerate(matrix1[ind]))
  scores1 = sorted(scores1, key=lambda x:x[1], reverse = True)

  scores2 = list(enumerate(matrix2[ind]))
  scores2 = sorted(scores2, key=lambda x:x[1], reverse = True)

  combined_scores = [(i,(s1+s2)/2) for (i,s1), (i,s2) in zip(scores1,scores2)]

  combined_scores = combined_scores[1:11]

  anime_indices = [k[0] for k in combined_scores]

  return df['Name'].iloc[anime_indices]
```

```python
recommendations_modified('One Punch Man', genre_similarity, plot_similarity)
```

```python
ratings = pd.read_csv("drive/MyDrive/Projects/Anime Recommendation Project/Datasets/rating_complete.csv")
```

```python
ratings.replace({-1: np.nan}, regex=True, inplace = True)
ratings.head()
```

```python
df = pd.read_csv("drive/MyDrive/Projects/Anime Recommendation Project/Datasets/anime.csv")
tvanime = df[df['type']=='TV']
tvanime.head()
```

```python
merged = ratings.merge(tvanime, left_on = 'anime_id', right_on = 'anime_id', suffixes= ['_user', ''])
```

```python
merged_new = merged[['user_id','name','rating_user']]
merged_new= merged_new[merged_new.user_id <= 100000]
```

```python
print(merged_new)
```

```python
pivot_table = merged_new.pivot_table(index = ["user_id"],columns = ["name"],values = "rating_user")
```

```python
anime_sample = pivot_table["One Punch Man"]
anime_similarity = pivot_table.corrwith(anime_sample)
anime_similarity = anime_similarity.sort_values(ascending=False)
anime_similarity.head()
```
